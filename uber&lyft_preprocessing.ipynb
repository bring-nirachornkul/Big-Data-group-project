```python
import re
def trunk_minisec(d):
  time_list = re.split('-| |:|\.', str(d))
  return time_list[0]+'-'+time_list[1]+'-'+time_list[2]+' '+time_list[3]+':'+time_list[4]+':'+time_list[5]
```

```python
df['time_stamp'] = pd.to_datetime(df['time_stamp']).apply(lambda x: trunk_minisec(x))
df = df.drop(['id', 'product_id'], axis=1)
```
