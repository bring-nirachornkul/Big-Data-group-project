```python
# why B02682 increase a lot

def get_tlc_rides(k, n):
  loc_by_m = [df_by_m[i][df_by_m[i]['Dispatching_base_num'] == tlc_code_2015[n]]['locationID'].values for i in range(6)]

  max_loc_by_m = []#location trips change by month
  max_trip_by_m = []#max k trips' amount
  loc_trip_dict = {}#max k trips' location
  for i in range(6):
    loc = df_by_m[i][df_by_m[i]['Dispatching_base_num'] == tlc_code_2015[n]]['locationID'].values
    loc_cnt = list(np.bincount(loc))
    loc_cnt_max = sorted(loc_cnt)[-k:] # most popular location in amount of trips
    m_popu = [loc_cnt.index(j) for j in loc_cnt_max] # most popular location
    max_loc_by_m.append(m_popu)
    max_trip_by_m.append(loc_cnt_max)
    for j in range(k):
      if max_loc_by_m[i][j] not in loc_trip_dict:
        loc_trip_dict[max_loc_by_m[i][j]] = [0]*6
      loc_trip_dict[max_loc_by_m[i][j]][i] = max_trip_by_m[i][j]
  
  return max_loc_by_m, max_trip_by_m, loc_trip_dict
```

```python
f_agg = ['other-FHV-services_jan-aug-2015.csv', 'Uber-Jan-Feb-FOIL.csv']
df_agg=[]
for i in f_agg:
    df_agg.append(pd.read_csv(path+i, encoding='Iso-8859-1'))
```

```python


df_tmp = df_agg[1].drop('trips', 1)

d = []
m = []
tmp_date = df_tmp['date'].values
for i in tmp_date:
  res = re.split('/', i)
  d.append(int(res[1]))
  m.append(int(res[0]))
df_tmp['Day'] = d
df_tmp['Month'] = m

df_tmp
```

```python
tlc_code_janfeb_tmp = ['B02512', 'B02598', 'B02617', 'B02682', 'B02764', 'B02765']

df_2015_wo_tmp = df_2015_wo.drop(df_2015_wo[df_2015_wo.Month>2].index)

ds_d_trips = df_2015_wo_tmp.groupby(['Dispatching_base_num','Month', 'Day']).size()
ds_h_trips = df_2015_wo_tmp.groupby(['Dispatching_base_num','Month', 'Day', 'Hour']).size()

```

```python
portion_by_h = []
for i in tlc_code_janfeb_tmp:
  for j in range(1, 3):
    for k in range(1, 32):
      if j == 2 and k == 29: break
      for l in range(0, 24):
        if l not in ds_h_trips[i][j][k]:
          tmp = 0
        else:
          tmp = round(ds_h_trips[i][j][k][l]/ds_d_trips[i][j][k], 2)
        portion_by_h.append(tmp)
```

```python
#TBD something wrong with data size
print(portion_by_h)
print(ds_h_trips)
print(act_car_by_h.shape)
```

```python
act_car_by_m = []
act_car_by_h = []
#company - month - day: 6*59=354
for i in tlc_code_janfeb_tmp:
  act_car_by_m += list(df_tmp[df_tmp['dispatching_base_number'] == i]['active_vehicles'].values)

#act_car_by_h: active car, company - month - day - hour, total 6*59*24 = 8496
for i in range(0, 354):
  for j in range(0, 24):
    act_car_by_h.append(int(round(act_car_by_m[i]*portion_by_h[i*24+j])))

```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor

from sklearn.metrics import r2_score

#input value
tlc_code_janfeb = np.array(['B02512']*59*24+['B02598']*59*24+['B02617']*59*24+['B02682']*59*24+['B02764']*59*24+['B02765']*59*24)
tlc_code_janfeb = tlc_code_janfeb.reshape(len(tlc_code_janfeb), 1)
week_janfeb = ['Thu']*24+['Fri']*24+['Sat']*24+['Sun']*24+['Mon']*24+['Tue']*24+['Wed']*24
week_janfeb = week_janfeb*8 + ['Thu']*24+['Fri']*24+['Sat']*24
week_janfeb = np.array(week_janfeb*6)
week_janfeb = week_janfeb.reshape(len(week_janfeb), 1)
hour_janfeb = np.array([i for i in range(0, 24)]*6*59)
hour_janfeb = hour_janfeb.reshape(len(hour_janfeb), 1)

#target value
act_car_by_h = np.array(act_car_by_h)
act_car_by_h = act_car_by_h.reshape(len(act_car_by_h), 1)

#assign as x, y data 
data_x = np.hstack((tlc_code_janfeb, week_janfeb, hour_janfeb))
data_y = act_car_by_h

#transform categorical values to one hot values
ct0 = ColumnTransformer(transformers=[('encoder0', OneHotEncoder(), [0])], remainder='passthrough')
ct1 = ColumnTransformer(transformers=[('encoder1', OneHotEncoder(), [6])], remainder='passthrough')
x0 = ct0.fit_transform(data_x)
x1 = ct1.fit_transform(x0).astype(float)

```

```python
#split data to 80, 20
train_x, test_x, train_y, test_y = train_test_split(x1, data_y, test_size = .2, random_state = 0)

#scale data for SVR
sc_x = StandardScaler()
sc_y = StandardScaler()
train_scx = sc_x.fit_transform(train_x)
train_scy = sc_y.fit_transform(train_y)

#create model
models=[]
models.append(('rbf SVR', SVR(kernel='rbf')))
models.append(('Random Forest', RandomForestRegressor(n_estimators=10, random_state=0)))
models.append(('multiple regression', LinearRegression()))
models.append(('decision tree', DecisionTreeRegressor(random_state=0)))


#train model
regressor = []
for i in range(len(models)):
  print("training {0} model...".format(models[i][0]))
  regressor.append(models[i][1])
  if i == 0:
    regressor[i].fit(train_scx, train_scy)
  else:
    regressor[i].fit(train_x, train_y)

#predict test data
accuracy = []

for i in range(len(models)):
  if i == 0:
    pred_y = regressor[i].predict(test_x)
    pred_y = sc_y.inverse_transform(regressor[i].predict(sc_x.transform(test_x)))
  else:
    pred_y = regressor[i].predict(test_x)
  print("model {0}, score: {1}".format(models[i][0], r2_score(test_y, pred_y)))
  accuracy.append(r2_score(test_y, pred_y))

print('\n', models[accuracy.index(max(accuracy))][0], "is the best")
```
